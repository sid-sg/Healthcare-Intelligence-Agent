# Load Data

```python
filepath = "/Volumes/workspace/default/hackathon/geocoded_dataset.csv"

df = spark.read \
    .option("header", True) \
    .option("multiLine", True) \
    .option("escape", '"') \
    .option("quote", '"') \
    .option("inferSchema", False) \
    .csv(filepath)
```

# Modify Data

## Remove unwanted columns

```python
df = df.drop(
    "pk_unique_id",
    "mongo DB",
    "unique_id",
    "content_table_id",
    "logo",
    "facebookLink",
    "twitterLink",
    "linkedinLink",
    "instagramLink",
    "missionStatementLink"
)
```

## Add primary key of table

```python
from pyspark.sql.functions import expr

df = df.withColumn("facility_id", expr("uuid()"))
```

## Fix data types

```python
from pyspark.sql.functions import col

df = df.withColumn("lat", col("lat").cast("double")) \
       .withColumn("lon", col("lon").cast("double")) \
       .withColumn("numberDoctors", col("numberDoctors").cast("double")) \
       .withColumn("capacity", col("capacity").cast("double")) \
       .withColumn("yearEstablished", col("yearEstablished").cast("double")) \
       .withColumn("acceptsVolunteers", col("acceptsVolunteers").cast("boolean"))
```

## Reorder table

```python
df.select(
    "facility_id",
    "name",
    "source_url",
    "specialties",
    "procedure",
    "equipment",
    "capability",
    "organization_type",
    "phone_numbers",
    "email",
    "websites",
    "officialWebsite",
    "yearEstablished",

    "acceptsVolunteers",
    "missionStatement",
    "organizationDescription",
    "facilityTypeId",
    "operatorTypeId",
    "affiliationTypeIds",
    "description",
    "area",
    "numberDoctors",
    "capacity",
    
    "address_line1",
    "address_line2",
    "address_line3",
    "address_city",
    "address_stateOrRegion",
    "address_zipOrPostcode",
    "address_country",
    "address_countryCode",
    "countries",

    "lat",
    "lon",
    "location_query",
    "osm_display_name",
    "method_used_for_geocoding",
    "resolved_location_query"
)
```

# Save Table

```python
df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.default.facilities")
```
